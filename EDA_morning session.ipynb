```python
#one-hot enoding ----> Creates multiple binary columns to represent each category
#Replaces numbers with 1s and 0s depending on the column value
#Eliminates any ordinal relationships between categories
#Can be used when the order of categorical features doesn't matter
```

```python

# Create dummy variables for 'Sex' column
o = pd.get_dummies(df, columns=['Sex'])

# Calculate correlation between the dummy 'Sex_female' and 'Age'
c_female = o['Sex_female'].corr(o['Age'])
c_male = o['Sex_male'].corr(o['Age'])

# Output the correlations
c_female, c_male
```
